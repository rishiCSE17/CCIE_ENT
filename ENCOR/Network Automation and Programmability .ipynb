## Guest Shell

* Guestshell is a __virtualized Linux-based environment__, designed to run custom Linux applications, including Python for automated control and management of Cisco devices. 
* The environment is intended for tools, Linux utilities, and manageability rather than networking
* Shares the kernel with the host (Cisco switches and routers) system.
* Guest Shell container is managed using IOx which is Cisco's Application Hosting Infrastructure for Cisco IOS. 
* Runs on an isolated memory segment to secure the IOS codes.
* During Guest Shell installation SSH is setup for key based authentication 
* User runs command with root priviledge 

### Guest Shell vs Guest Shell Lite

||Guest Shell Lite | Guest Shell |
|--|--|--|
|Operating System| IoS XE | IOS XE |
|Shell Environment| Montavista CGE7 | Centos7 |
|Python version| 2.7.11 | 2.7.5 |
|Custom libs | Cisco EEM <br> XE CLI <br> ncclient | Cisco EEM <br> XE CLI |
|Root FS | Busybox, SSH, Pip | SSH, Yum, Pip |
|RPM install | No | yes |
|Architecture | MIPS | x86 |

### IOx

* IOx is a Cisco-developed end-to-end application framework that provides application hosting capabilities for different application types on Cisco network platforms. 
* App hosting provides the following features
    * Hides network heterogeneity.
    * IOx application programming interfaces (APIs), remotely manage the life cycle of applications hosted on a device.
    * Centralized app life-cycle management.
    * Cloud-based developer experience.
    
### Guest-shell Activation

```
conf t
    iox                           !enable iox platform
    exit
    app-hosting appid guestshell  ! activate guest-shell interface
        vnic management guest-interface 0
end
guestshell enable                 ! activate huest shell
guestshell run [CMD|bash]         ! running command or bash shell
```

__Verification__

```
show iox-service
show app-hosting list
```

__Stop service__

```
guestshell disable
guestshell destroy
```
